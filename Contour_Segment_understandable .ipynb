# This code is used to draw contours on an image as a technique for segmentation

```python
from skimage.color import rgb2gray
from skimage import measure
from skimage import io
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
```

Upload image and convert it to grayscale and see if image is uploaded

im = io.imread('dtest2.jpg',as_gray=True); 
image = rgb2gray(im)
plt.imshow(image,cmap='gray');

Use Otsu to binarize and add threshold to image

```python
#Getting threshold value
thresh = threshold_otsu(image)
#Applying threshold
thresholded_image = image > thresh
```

Find contours at a certain value

```python
contours = measure.find_contours(thresholded_image, 0.8)
```

We can use the next two lines to determine the number of dots and calculate the total score. The dimensions are (x,y,number of dots). The spatial coordinates match the dimensions of the resulting picture. Silence the next two lines if necessary.

```python
#for contour in contours:
    #print(contour.shape)
```

```python
fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)
for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()   
```
